### Objetivo: Comparar uma carteira aleatória de ações brasileiras com o desempenho do BOVA11 em um período de tempo:

- 1º - Importação de bibliotecas necessárias:

```python
from pandas_datareader import data as web
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import random
```

- 2º - Definição do período para análise:

```python
inicio = '2020-01-01'
fim = '2022-11-01'
```

- 3º - Importando dados da planilha com ativos que compõem o Ibovespa fornecida pelo site da B3: https://www.b3.com.br/pt_br/market-data-e-indices/indices/indices-amplos/indice-ibovespa-ibovespa-composicao-da-carteira.htm

```python
ibovespa = pd.read_excel("IBOVDia.xlsx")
ibovespa
```

- 4º - Selecionando apenas a coluna dos 'Tickers' das ações:

```python
ibovespa['Código']
```

- 5º - Adequando todos os códigos para captura de dados de fechamento na biblioteca yfinance:

```python
tickers = []
for i in ibovespa['Código']:
    tickers.append(i+'.SA')
tickers      
```

- 6º - Utilizando a função 'random' , escolhemos 10 ativos aleatórios da lista de tickers:

```python
sort = random.sample(tickers,10)
sort
```

- 7º criando um dataframe que irá conter a leitura dos dados retirados da biblioteca yf, contendo os parâmetros estipulados:

```python
df_carteira = web.DataReader(sort, data_source = 'yahoo', start = inicio , end= fim)['Adj Close']
df_carteira
```

- 8º plotagem de um gráfico de acordo com dados coletados (dimensões do gráfico alteradas para melhor visualização):

```python
df_carteira.plot(figsize=(15,10));
```

- 9º - Selecionando os dados do dataframe da primeira linha até o final -> método .iloc
---
- 10º - Normalizando os dados para análise de retorno acumulado, partindo de um mesmo ponto.

```python
df_carteira.iloc[0]
carteira_normalizada = df_carteira/df_carteira.iloc[0]
carteira_normalizada
```

```python
carteira_normalizada['Carteira'] = carteira_normalizada.sum(axis=1)
carteira_normalizada
```

```python
df_bova = web.DataReader('BOVA11.SA', data_source = 'yahoo', start = inicio , end= fim)['Adj Close']
df_bova
```

```python
compar = pd.merge(carteira_normalizada, df_bova, how = "inner", on = 'Date')
compar
```

```python
compar.rename(columns = {'Adj Close':'BOVA11.SA'}, inplace = True)
compar
```

```python
compar.iloc[0]
retorno_comparado = compar / compar.iloc[0]
retorno_comparado
```

```python
retorno_comparado[['Carteira','BOVA11.SA']].plot(figsize=(15,10));
```

#### Observações:
    - Dependendo o período que se estipula para busca de dados, alguns ativos podem aparecer como resultado 'NaN' na coluna de Fechamento Ajustado, isto pode acontecer caso a base de dados não esteja atualizada e o ativo tenha mudado de nome, ou também o ativo pode não estar listado em bolsa no período exigido.
